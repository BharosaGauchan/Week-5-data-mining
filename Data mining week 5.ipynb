### 1. Clean the “Tweet Content” column by removing non-text data and stop words

```python
# reading file
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\bharo\Downloads\raw.githubusercontent.com_bellevue-university_dsc360_main_12 Week_week_5_data_twitter_sample.csv")
df.head()  
```

```python
# removing special characters
df["Tweet Content"] = df['Tweet Content'].str.replace("[^\w\s]"," ")
```

```python
#Dropping stop words using nltk package
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_word = stopwords.words("english")
df["Tweet Content"] = df["Tweet Content"].apply(lambda x: " ".join([word for word in str(x).split() if word not in stop_word]))
df.head()
```
